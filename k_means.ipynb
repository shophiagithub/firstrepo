EMPLOYEES SALARY PREDICTION BASED ON EMPLOYEES AGE USING K-MEANS ALGORITHM

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("income.csv")
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
# creating an object
km=KMeans(n_clusters=3)
km
```

```python
y_predict=km.fit_predict(df[['Age','Income($)']])
y_predict
```

```python
df['cluster']=y_predict
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1['Age'],df1['Income($)'],color='green')
plt.scatter(df2['Age'],df2['Income($)'],color='red')
plt.scatter(df3['Age'],df3['Income($)'],color='black')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend
```

```python
#creating object
scaler=MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)']=scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
df
```

```python
km=KMeans(n_clusters=3)
y_predict=km.fit_predict(df[['Age','Income($)']])
y_predict
```

```python
df['cluster']=y_predict
df
```

```python
#printing the clusters
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1['Age'],df1['Income($)'],color='green')
plt.scatter(df2['Age'],df2['Income($)'],color='red')
plt.scatter(df3['Age'],df3['Income($)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend
```

```python
#randomly iniatializing k using elbow method
k_range=range(1,10)
cost_j=[]
for k in k_range:
  km=KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  cost_j.append(km.inertia_)# inertia is nothing but the computed cost function of this model where all the values are appended
  # to the  cost_j array and a line plot is plotted clusters vs cost_j to choose the correct value of k
```

```python
cost_j
```

```python
plt.xlabel('K cluster')
plt.ylabel('cost function')
plt.plot(k_range,cost_j)
```
